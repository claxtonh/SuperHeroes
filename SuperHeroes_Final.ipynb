```python
import pandas as pd
from sqlalchemy import create_engine
import requests
import json
import pymongo
```

```python
csv_file_info2 = "marvel_characters_info.csv"
hero_data_info2 = pd.read_csv(csv_file_info2)
hero_data_info2.head()
csv_file_powers = "superheroes_power_matrix.csv"
hero_powers = pd.read_csv(csv_file_powers)
hero_powers.head()
```

```python
hero_powers.rename(columns={"Name": "SuperHero Name"}, inplace=True)
hero_powers.count()
```

```python
hero_data_info2.rename(columns={"Name": "SuperHero Name"}, inplace=True)
hero_data_info2.count()
```

```python
combined = pd.merge(hero_data_info2, hero_powers, how="left", on="SuperHero Name")
combined.head()
combined.count()
```

```python
combined = combined.drop(["ID", "EyeColor", "Race", "HairColor", "SkinColor"], axis=1)
```

```python
combined.head()
```

```python
combined.Publisher.unique()
#According to Google, the Top 5 comicbook publishers are Marvel, DC Comics, Image Comics, IDW Publishing and Dark Horse Comics
```

```python
#Connecting the MongoDatabase
conn = "mongodb://localhost:27017"
client = pymongo.MongoClient(conn)
```

```python
#Creating the database Comic_SuperHeroes
db = client.Comic_SuperHeroes
```

```python
collection = db.Marvel_SuperHeroes
Marvel_only = combined[combined.Publisher == "Marvel Comics"]
Marvel_only.count()
df = Marvel_only
records_ = df.to_dict(orient = 'records')
result = db.Marvel_SuperHeroes.insert_many(records_)
#result
```
